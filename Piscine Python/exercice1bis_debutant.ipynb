```python
#exercice 2
#Écrire dans une cellule un programme qui demande à l'utilisateur de rentrer un rayon et la plus haute, puis qui calcule le volume d'un cône droit. 
# (Rappel : Vcone = 1/3 * pi * r^2 * h) Bis : mettre ce programme dans une fonction volume_cone() et tester cette fonction

def volume_cone(r, h):
    return (1/3) * 3.14 * r**2 * h

rayon = float(input("Saisissez le rayon de la base du cône (en cm) : "))
hauteur = float(input("Saisissez la hauteur du cône (en cm) : "))

volume = volume_cone(rayon, hauteur)

print(f"V = {volume} cm3")
```

```python
#exercice 3
#Écrire dans une cellule un programme qui demande à l'utilisateur son poids et sa taille et qui calcule son IMC (trouver la formule sur internet).
#Bis : mettre ce programme dans une fonction imc() et tester cette fonction

def imc(p, t):
    return p / t**2

poids = float(input("Saisissez votre poids (en kg) : "))
taille = float(input("Saisissez votre taille (en m) : "))

votre_imc = imc(poids, taille)

print(f"Votre imc est de {votre_imc}.")

if votre_imc < 18.5:
    print("Vous êtes en insuffisance pondérale.")
elif votre_imc < 25:
    print("Votre poids est normal.")
elif votre_imc < 30:
    print("Vous présentez des signes de surpoids.")
elif votre_imc < 35:
    print("Vous êtes en obésité de niveau 1.")
elif votre_imc < 40:
    print("Vous êtes en obésité de niveau 2.")
else:
    print("Vous êtes en obésité de niveau 3.")
```

```python
#exercice 4
#Écrire une fonction qui convertit en mètres par seconde et en km/h une vitesse fournie en paramètres par l'utilisateur en miles/heure ( 1 mile = 1609 mètres ).
# Ce résultat sera retourné par la fonction.
def conv_vitesse(v):
    conv_km_h = v/1.609
    conv_m_s = conv_km_h/3.6
    return conv_km_h, conv_m_s

vitesse_miles_h = float(input("Saisissez une vitesse en miles/h : "))

vitesse_km_h, vitesse_metre_s = conv_vitesse(vitesse_miles_h)

print(f"{vitesse_miles_h} miles/h vaut {vitesse_km_h} km/h, soit {vitesse_metre_s} m/s.")
```

```python
#exercice 6
#Écrire un programme en python pour réviser ses tables de multiplication. 
#Le programme tire au hasard deux entiers entre 0 et 10 et demande à l'utilisateur de donner leur produit.
#Si l'utilisateur répond juste, il gagne un point. S'il répond faux, il perd un point.
#On interrogera 5 fois l'utilisateur.
#Le programme affiche à la fin le score de l'utilisateur.

import random

score = 0

for i in range(0, 5):
    x = random.randint(0, 10)
    y = random.randint(0, 10)
    produit = int(input(f"{x} x {y} =  ? "))
    if produit == x*y:
        score += 1
    else:
        score -= 1

print(f"Votre score est de {score}.")
```

```python
#exercice 7
#L'ordinateur tire un nombre entier au hasard entre 0 et 100.
#L'utilisateur doit le trouver et pour cela proposer des valeurs.
#L'ordinateur indique pour chaque valeur proposée si la valeur est trop petite, trop grande ou s'il a trouvé.
#a) Écrire un programme en Python pour jouer à ce jeu.
#b) Modifier le programme pour qu'il s'arrête si l'utilisateur n'a pas trouvé au bout de 10 coups.
#c) Mettre ce programme dans une fonction plus_ou_moins() qui prendra en paramètres :
#   - le nombre min et max de la plage de tirage du nombre au hasard
#   - le nombre de coups maximum

def le_juste_nombre(min, max, essaie):
    solution = random.randint(min, max)
    guess = -1
    coup = 0
    victoire = False
    while coup < essaie and guess != solution:
        coup += 1
        guess = int(input("Devinez le nombre : "))
        if guess == solution:
            print("Bravo ! ")
            victoire = True
        elif guess < solution:
            print("C'est plus ! ")
        elif guess > solution:
            print("C'est moins ! ")
    
    if victoire:
        print(f"Vous avez gagné ! La solution était bien {solution} ! ")
    else:
        print(f"Dommage ! La solution était {solution}.")

minimum = int(input("Entrez un nombre positif minimum : "))
maximum = int(input("Entrez un nombre positif maximum : "))
nb_essaie = int(input("Combien d'essaie voulez-vous ? "))

le_juste_nombre(minimum, maximum, nb_essaie)
```

```python
#exercice 8
#Ecrire un programme qui tourne indéfiniment et qui demande à l'utilisateur de rentrer un nombre entier compris dans une plage définie aléatoirement 
# (par exemple, entrer un nombre compris entre 12 et 34).
#Si le nombre entré est en dehors de la plage, le programme l'affiche et redemande à nouveau un nombre dans la même plage.
#Si le nombre entré est dans la plage, on félicite l'utilisateur et le programme reprend avec une nouvelle plage.

def intervalle(min, max):
    entre_deux = int(input(f"Entrez un nombre entre {min} et {max} : "))
    while entre_deux < min or entre_deux > max:
        entre_deux = int(input(f"Non ! Saisissez un nombre entre {min} et {max} : "))
    print(f"Oui ! {entre_deux} est compris entre {min} et {max} ! ")

while True:
    x = random.randint(0, 100)
    y = random.randint(0, 100)
    if x < y:
        intervalle(x, y)
    else:
        intervalle(y, x)
```

```python
#exerice 9
#Écrire un programme permettant d'entrer des notes d'élèves sur 20 jusqu'à ce que l'utilisateur saisisse une note vide. 
#Cette fonction retourne le nombre de notes entrées, la note la plus élevée, la note la plus basse, la moyenne de toutes les notes.
note_chaine = "0"
liste_notes = []
nb_notes = 0
max = 0
min = 20
total = 0

while note_chaine != "":
    note_chaine = input("Saisissez une note (sur 20) : ")
    if note_chaine != "":
        note_nombre = float(note_chaine)
        liste_notes.append(note_nombre)
        nb_notes += 1

for i in liste_notes:
    total += i
    if i >= max:
        max = i
    if i <= min:
        min = i

moyenne = total/nb_notes

print(f"Sur les {nb_notes} notes saisies, la plus élevée est de {max}/20, la plus basse est de {min}/20, et la moyenne est de {moyenne}.")
```
