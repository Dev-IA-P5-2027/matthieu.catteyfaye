```python
# 2. Écrire une fonction qui retourne f(x) = 2x3 + x − 5. Écrire une fonction poly avec quatre paramètres : fonction, borneInf, borneSup et nbPas. 
# Cette procédure affiche les valeurs de fonction, de borneInf à borneSup, touslesnbPas. Elle doit respecter borne Inf < borne Sup.
# La fonction poly doit renvoyer un dictionnaire clé:valeur (clé = x, valeur = f(x)).

def fonction_degre3(x):
    return 2 * x**3 + x -5

def poly(fonction, borne_inf, borne_sup, nb_pas):
    dico = {}
    if borne_inf < borne_sup:
        for x in range(borne_inf, borne_sup+1, nb_pas):
            dico[x] = fonction(x)
    else:
        print(f"Erreur : {borne_sup} < {borne_inf}")
    return dico

borne_min = int(input("Saisissez la borne inférieure : "))
borne_max = int(input("Saisissez la borne supérieure : "))
valeur_pas = int(input("Saisissez le pas : "))

valeurs_fonction_degre3 = poly(fonction_degre3, borne_min, borne_max, valeur_pas)
print(valeurs_fonction_degre3)
```

```python
# 4. Écrire une fonction compterMots ayant un argument (une chaîne de caractères) 
# et qui renvoie un dictionnaire qui contient la fréquence de tous les mots de la chaîne d'entrée.


# ne marche qu'avec des phrases sans ponctuation
def compterMots(phrase:str)->dict:
    frequence_mots = {}
    liste_mots = phrase.split(" ")
    for i in liste_mots:
        if (i in frequence_mots):
            frequence_mots[i] += 1
        elif i != '':
            frequence_mots[i] = 1
    return frequence_mots

votre_phrase = input("Saisissez une phrase : ")
dico_frequence_mots = compterMots(votre_phrase)
print(dico_frequence_mots)
```

```python
# 5. Fonction renvoyant plusieurs valeurs sous forme d’un tuple. 
# Écrire une fonction minMaxMoy qui reçoit une liste d’entiers et qui renvoie le minimum, le maximum et la moyenne de cette liste. 
# Le programme principal appellera cette fonction avec la liste : [10, 18, 14, 20, 12, 16].

def minmaxmoy(l):
    max = 0
    min = 20
    total = 0
    for i in l:
        if i >= max:
            max = i
        if i <= min:
            min = i
        total += i
    moy = total/len(l)
    return (min, max, moy)

liste_notes = [10, 18, 14, 20, 12, 16]

tuple_minmaxmoy = minmaxmoy(liste_notes)
print(tuple_minmaxmoy)
```

```python
# 7. Le but de cet exercice est de créer une fonction à laquelle on donne un message et un décalage, et la fonction renvoie alors le message décalé dans l'alphabet. 
# Il faudra faire attention que le message peut contenir des caractères ne faisant pas forcément partie de l'alphabet et dans ce cas, pour ne pas perdre la signification du texte, 
# ces caractères doivent réapparaitre à l'identique dans le message chiffré. De plus, il faudra gérer le dépassement 
# ('z' décalé vers la droite revient sur 'a', et 'a' décalé vers la gauche revient sur 'z').

# À noter que la fonction pourra être utilisée aussi bien pour chiffrer que pour déchiffrer (il suffit pour cela de lui passer le message chiffré avec l'opposé du décalage utilisé 
# pour retrouver le message d'origine).


def cryptage(message:str, decalage:int):
    message_code = ""
    nouv_decalage = decalage%26
    for i in message:
        if 65<=ord(i)<=90:
            nouvel_ascii = ord(i)+nouv_decalage
            if nouvel_ascii < 65:
                nouvel_ascii += 26
            elif nouvel_ascii > 90:
                nouvel_ascii -= 26
            nouveau_char = chr(nouvel_ascii)
            message_code += nouveau_char
        
        elif 97<=ord(i)<=122:
            nouvel_ascii = ord(i)+nouv_decalage
            if nouvel_ascii < 97:
                nouvel_ascii += 26
            elif nouvel_ascii > 122:
                nouvel_ascii -= 26
            nouveau_char = chr(nouvel_ascii)
            message_code += nouveau_char
            
        else:
            message_code += i
    return message_code

votre_message = input("Saisissez le message à coder/décoder : ")
votre_decalage = int(input("Saisissez le décalage : "))

votre_message_code = cryptage(votre_message, votre_decalage)
print(votre_message_code)
```
