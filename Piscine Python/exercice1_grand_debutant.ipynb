```python
#exercice 17
mot = input("Saisissez un mot : ")
compteur = 0
mot_unique = []
#conversion de la string en liste de caractères uniques
for x in mot:
    if x not in mot_unique:
        mot_unique.append(x)

#print(mot_unique)

for i in mot_unique:
    for j in mot:
        if i == j:
            compteur += 1
    print(f"{i} figure {compteur} fois.")
    compteur = 0
```

```python
#exercice 20
chaine = input("Saisissez une chaîne : ")
conv_liste = []

for i in chaine:
    conv_liste.append(i)

conv_liste[0], conv_liste[-1] = conv_liste[-1], conv_liste[0]

#Je reconvertis en chaine avec join : je mets dans une même string les éléments de conv_liste
chaine = "".join(conv_liste)

print(chaine)
```

```python
#exercice 22
chaine = input("Saisissez une phrase : ")
itteration = 0
conv_liste = []
mot_isole = []

#conversion de la string en liste
for i in chaine:
    conv_liste.append(i)

char = conv_liste[0]

#Je récupère les caractères de la string jusqu'à la fin de la chaine ou jusqu'à la fin du premier mot
while itteration < len(chaine) and char != " " and char != "," and char != ";" and char != "." and char != ":" and char != "!" and char != "?":
    mot_isole.append(char)
    itteration += 1
    char = conv_liste[itteration]

mot_isole_chaine = "".join(mot_isole)
print(mot_isole_chaine)
```

```python
#exercice 24
chaine = input("Saisissez un mot : ")
liste_conv = []
check_palindrome = True
index_liste = -1

for i in chaine:
    liste_conv.append(i)

for i in liste_conv:
    if check_palindrome == True:
        if i == liste_conv[index_liste]:
            check_palindrome = True
        else:
            check_palindrome = False
    index_liste -= 1

if check_palindrome:
    print(f"{chaine} est un palindrome ! ")
else:
    print(f"{chaine} n'est pas un palindrome.")
```
