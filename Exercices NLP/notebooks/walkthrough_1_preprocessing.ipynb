```python
# 2. Handling Contractions
import contractions

def fix_contractions(data):
    expanded_corpus = [contractions.fix(text) for text in data]
    return expanded_corpus
```

```python
corpus = [
    "I can't wait for the new season of my favorite show!",
    "Lemme know when you're ready.",
    "Let's go!",
    "Nothing is gonna change my love for you"
]

fix_contractions(corpus)
```

```python
# 3. Spell Checking
from autocorrect import Speller

def correct_with_autocorrect(text, lang='en'):
    """Correct spelling using autocorrect library for English"""
    spell = Speller(lang=lang)
    return spell(text)

def correct_with_autocorrect_fr(text, lang='fr'):
    """Correct spelling using autocorrect library for French"""
    spell = Speller(lang=lang)
    return spell(text)
```

**3. Standardizing Accent Characters**

```python
import unicodedata

def standardize_accented_chars(text):
    return unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
```

```python
input_text = 'Sóme words such as résumé, café, prótest, divorcé, coördinate, exposé, latté.'
print(standardize_accented_chars(text=input_text))
```

```python
# 1. Removing html tags

def remove_htlm_tags_regex(text):
    """Remove HTML tags using regular explression."""
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text) # On change l'expression en '' (vide).
```

```python
html_text= "<p>Hello, <b>world</b>!</p> <a href='#'>Click here</a>"
clean_text = remove_htlm_tags_regex(html_text)
print(clean_text)
```

```python
# 2. Removing numbers/digits
def remove_numbers(text):
    result = re.sub(r'\d+', '<phone number>', text) # '\d' pour 'digit' et '+' pour indiquer les nombres valant 1 ou plus. On peut aussi utiliser '\d{7,}' pour les nombres valant au moins 7.
    return result

def remove_numbers_seven(text):
    result = re.sub(r'\d{7,}', '<phone number>', text) 
    return result
```

```python
input_text = "Her phone number is 0123456789, and my phone number is 0987654321"
remove_numbers(input_text)
```

```python
remove_numbers_seven(input_text)
```

```python
# 3. Convert the numbers into words
import inflect
p = inflect.engine()

# convert into words
def convert_number(text):
    # split string into list of words
    temp_str = text.split()
    # initialise empty list
    new_string = []

    for word in temp_str:
        # if word is a digit, convert the digit
        # to numbers and append into the new_string list
        if word.isdigit():
            temp = p.number_to_words(word)
            new_string.append(temp)

        # append a word as it is
        else:
            new_string.append(word)

    # join the words of the new_string to form a string
    temp_str = ' '.join(new_string)
    return temp_str

input_text = "There are 3 balls in this bag, and 12054465496 in the other one."
convert_number(input_text)
```

```python
# 4. Removing Punctuation & Special Characters
regular_punct = list(string.punctuation)
print(regular_punct)
```

```python
def remove_punctuation(text, punct_list):
    for punc in punct_list:
        if punc in text:
            text = text.replace(punc, '')
    return text.strip()

remove_punctuation("Hey @Lisa, did you know that the summer break is comming? Amazing right !! It's only 5 more days !!", regular_punct)
```

```python
#6. Handling emoji/emoticon ---> for sentiment analysis, emoji/emoticon is usually not removed

def remove_emojis_regex(text):
    """Remove emoji using Unicode emoji ranges."""
    # Unicode ranges for emojis
    emoji_pattern = re.compile("["
                              u"\U0001F600-\U0001F64F" # emoticons
                              u"\U0001F300-\U0001F5FF" # symbols & pictographs
                              u"\U0001F680-\U0001F6FF" # transport & map symbols
                              u"\U0001F1E0-\U0001F1FF" # flags (iOS)
                              u"\U00002702-\U000027B0" # dingbats
                              u"\U000024C2-\U0001F251"
                              u"\U0001f926-\U0001f937"
                              u"\U00010000-\U0010ffff"
                              "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
# Exemple
text = "Hello😄! How are you🍕? I'm in nyc🏴!"
clean_text = remove_emojis_regex(text)
print(clean_text)
```

```python
import emoji

def remove_emojis_emoji_library(text):
    """Remove emojis with their text description."""
    return emoji.replace_emoji(text, replace='')

def replace_emojis_with_text(text):
    """Replace emojis with their text description."""
    return emoji.demojize(text)

# Exemple
text = "I love Python!🐍❤️ Let's code!💻"

print(remove_emojis_emoji_library(text))
print(replace_emojis_with_text(text))
```

```python
# 7. Removal of URLs or other irrelevant hints

# Define a regex pattern to match URLs
url_pattern = re.compile(r"https?://\S+")

# Define a function to remove URLs from text
def remove_urls(text):
    return url_pattern.sub('', text) 

text = "Visit https://www.google.com for more information."
remove_urls(text)
```

```python
url_pattern=[]
url_pattern.append(re.compile(r'https:?//\S+'))
url_pattern.append(re.compile(r'http:?//\S+'))
url_pattern.append(re.compile(r'www.\S+'))

def remove_urls2(text):
    for pattern in url_pattern:
        text=pattern.sub('<this link>', text) # we replace the URLs with word "<this link>"
    return text

input_text="Visit the following link: www.google.com"
print(remove_urls2(input_text))
```

```python
# The requiered libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download the necessary NLTK datasets
nltk.download('punkt')
nltk.download('stop_words')

# remove stopwords function
def remove_stopwords(text):
    stop_words = set(stopwords.words("english"))
    word_tokens = word_tokenize(text)

    # Filter out the stopwords
    filtered_text = [word for word in word_tokens if word.lower() not in stop_words]

    return filtered_text

example_text = "This is a simple sentence and we are going to remove the stopwords from this."

print("Before removing stopwords:", example_text)
print("After removing stopwords:", remove_stopwords(example_text))
```

**7. Stemming & Lemmatization**

```python
#Stemming 

from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize
stemmer = PorterStemmer()

# stem words in the list of tokenized words
def stem_words(text):
    word_tokens = word_tokenize(text)
    stems = [stemmer.stem(word) for word in word_tokens]
    return stems

text = 'data science uses scientific methods algorithms and many types of processes'

stem_words(text)
```

```python
# Lemmatization using NLTK

from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download the necessary NLTK datasets
nltk.download('punkt')
nltk.download('wordnet')

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Function to lemmatize words in the list of tokenized words
def lemma_words(text):
    word_tokens = word_tokenize(text)
    lemmas = [lemmatizer.lemmatize(word) for word in word_tokens]
    return lemmas

# Exemple text
text = 'The quick brown foxes are jumping over the lazy dogs.'

print(lemma_words(text))
```

```python
# Lemmatization using spaCy
import spacy

# Load the spaCy English model
nlp = spacy.load('en_core_web_sm')

# Define a sample text
text = "The quick brown foxes are jumping over the lazy dogs."

# Process the text using spaCy
doc = nlp(text)

# Extract lemmatized tokens
lemmatized_tokens = [token.lemma_ for token in doc]

# Join the lemmatized tokens into a sentence
lemmatized_text = ' '.join(lemmatized_tokens)

# Print the original and lemmatizes text
print("Original Text:", text)
print("Lemmatized Text:", lemmatized_text)
```
