```python
import re

pattern = re.compile(r"""
   ^                     # start of line
   (\d{2})               # day
   -                     # separator
   (\d{2})               # month
   -                     # separator
   (\d{4})               # year
   $                     # end of line
""", re.VERBOSE)

match = pattern.match('12-01-2024')

if match:
   print('Day:', match.group(1))
   print('Month:', match.group(2))
   print('Year:', match.group(3))
```

```python
email_pattern = re.compile(r'''
    ^                   # Start of the line
    [a-zA-Z0-9._%+-]+   # Username
    @                   # @ symbol
    [a-zA-Z0-9.-]+      # Domain name
    \.                  # Dot
    [a-zA-Z]{2,}        # Top-level domain (at least 2 characters)
    $                   # End of the line
''', re.VERBOSE)

emails = [
    'example@email.com',
    'user.name@email.domain',
    '1234@invalid',
    '@missingusername.com',
    'email@subdomain.co.uk'
]

for email in emails:
    if email_pattern.match(email):
        print(f"{email}: Valid")
    else:
        print(f"{email}: Invalid")
```

```python
text = """This is the first line.
This is the second line.
This is the third line.
"""

# Without re.MULTILINE, ^ and $ match the start and end of the string
matches = re.findall('^This', text)
print(matches) # Output: ['This']

matches = re.findall('line.$', text)
print(matches) # Output: ['line.']

# With re.MULTILINE, ^ and $ match the start and end of each line
matches = re.findall('^This', text, re.MULTILINE)
print(matches) # Output: ['This', 'This', 'This']

matches = re.findall('line.$', text, re.MULTILINE)
print(matches) # Output: ['line.', 'line.', 'line.']
```
